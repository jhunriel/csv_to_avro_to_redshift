## Elbow Method

```python
wcss = []
max_cluster=15
for i in range(1,max_cluster):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
number_clusters = range(1,max_cluster)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,max_cluster))


visualizer.fit(x)        
visualizer.show()        
```

## Silhouette Method

```python
sil = []
for k in range(2, max_cluster):
  kmeans = KMeans(n_clusters = k).fit(x)
  labels = kmeans.labels_
  sil.append(silhouette_score(x, labels, metric = 'euclidean'))
```

```python
number_clusters = range(2,max_cluster)
plt.plot(number_clusters, sil)
plt.title('The Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
```

```python
model = KMeans(4)
```

```python
model.fit(x)
```

```python
clustered_data = data.copy()
x['cluster']  = model.fit_predict(x)
clustered_data['cluster']  = model.fit_predict(x)
```

```python
clustered_data.head(10)
```

```python
def predict(userid,count,df):
    userratings  = df.loc[userid]
    noratings = userratings[userratings.isnull()]
    withincluster = df[df.cluster==userratings.cluster].mean()
    recommended = pd.concat([noratings,withincluster],axis=1, join='inner').loc[:,0].sort_values(ascending=False)[:count]
    return recommended
```

```python
predict(124,12,clustered_data)
```
